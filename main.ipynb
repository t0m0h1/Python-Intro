# Introduction to Python

Python is a **high-level, interpreted programming language** known for its simplicity and readability. It is widely used for **web development, data analysis, artificial intelligence, automation, and more**.  

Some key features of Python include:

- **Easy to read and write:** Python uses clear, English-like syntax, making it beginner-friendly.
- **Interpreted:** You can run Python code directly without compiling it first.
- **Dynamically typed:** You don’t need to declare variable types; Python figures them out automatically.
- **Versatile:** Python supports multiple programming paradigms, including **procedural, object-oriented, and functional programming**.
- **Large standard library and community:** Python comes with many built-in modules, and there are countless external libraries available for specialized tasks.

In this notebook, we will start by exploring **Python’s core data structures** and how they are used to store and manipulate data efficiently.


## 1. Lists
A Python list is one of the most commonly used data structures. 

In simple terms:
- It's an ordered collection of items.
- It can hold any type of data: numbers, strings, booleans, or even other lists.
- It is mutable, which means you can change, add, or remove items after creating it.

Think of it like a row of boxes, where each box has a number (called an index) starting from 0, and you can put anything inside, take things out, or swap them around.


```python
# Creating a list
fruits = ["apple", "banana", "cherry"]

# Accessing elements
print(fruits[0])  # apple

# Adding elements
fruits.append("orange")

# Removing elements
fruits.remove("banana")

# Slicing
print(fruits[1:3])
```
